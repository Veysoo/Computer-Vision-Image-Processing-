#### HISTOGRAM

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("red_blue.png")
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(image)

print(image.shape)

img_hist = cv2.calcHist([image],channels=[0],mask=None,histSize=[256],ranges = [0,256])

plt.figure()
plt.plot(img_hist)
```

```python
color = ("b","g","r")
plt.figure()

for i , c in enumerate(color):
    hist = cv2.calcHist([image],channels=[i],mask=None,histSize=[256],ranges = [0,256])
    plt.plot(hist,color = c)
```

```python
golden_gate = cv2.imread("golden_gate.jpg")
golden_gate = cv2.cvtColor(golden_gate,cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(golden_gate)

print(golden_gate.shape)

mask = np.zeros(golden_gate.shape[:2],np.uint8)

plt.figure()
plt.imshow(mask,cmap="gray")
```

```python
mask[50:100,100:150] = 255

plt.figure()
plt.imshow(mask,cmap="gray")

masked_image_vis = cv2.bitwise_and(golden_gate,golden_gate,mask= mask)

plt.figure()
plt.imshow(masked_image_vis,cmap="gray")
```

```python
masked_img = cv2.bitwise_and(golden_gate,golden_gate,mask= mask)

masked_img_hist = cv2.calcHist([masked_img],channels=[0],histSize=[256],mask=mask,ranges=[0,256])
plt.figure()
plt.plot(masked_img_hist)
```

#### HISTGORAM ETLEME (konrtrast arttrmamz salyor)(kartlk arttrma)

```python
pic = cv2.imread("pict.jpg",0)
plt.figure(),plt.imshow(pic,cmap="gray")

pic_hist = cv2.calcHist([pic],channels=[0],histSize=[256],mask=None,ranges=[0,256])
plt.figure(),plt.plot(pic_hist)

#eitleme

eq_hist = cv2.equalizeHist(pic)
plt.figure(),plt.imshow(eq_hist,cmap="gray")

eq_img_hist = cv2.calcHist([eq_hist],channels=[0],histSize=[256],mask=None,ranges=[0,256])
plt.figure(),plt.plot(eq_img_hist)
```

THRESHOLDING

```python
import cv2
import numpy as numpy
import matplotlib.pyplot as plt

#image reading white-black

image = cv2.imread("desert.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

plt.figure()
plt.axis("off")
plt.imshow(image,cmap="gray")


# thresholding

_ , image_thresh = cv2.threshold(image,thresh=100,maxval=255, type= cv2.THRESH_BINARY)

plt.figure()
plt.axis("off")
plt.imshow(image_thresh,cmap="gray")


_ , image_thresh = cv2.threshold(image,thresh=100,maxval=255, type= cv2.THRESH_BINARY_INV)

plt.figure()
plt.axis("off")
plt.imshow(image_thresh,cmap="gray")



# Adaptive Thresholding

image_tresh_2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
#                                     (image , maxvalue , method , type , block_size )

plt.figure()
plt.axis("off")
plt.imshow(image_tresh_2,cmap="gray")
```

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt


# opencv default color is bgr . if we want to using rgb colors scales , we should transform to rgb  
# bgr 

img_desert = cv2.imread("desert.jpg")
img_beach = cv2.imread("beach.jpg")


# to rgb

img_desert = cv2.cvtColor(img_desert,cv2.COLOR_BGR2RGB)
img_beach = cv2.cvtColor(img_beach,cv2.COLOR_BGR2RGB)


#we have to resize image because it have to be same shape

print(img_beach.shape)
print(img_desert.shape)

#resize images

img_desert = cv2.resize(img_desert,(600,600))
img_beach = cv2.resize(img_beach,(600,600)) 

print(img_beach.shape)
print(img_desert.shape)

# blending  (image = alpha*image1 + beta*image2   ........)

img_blended = cv2.addWeighted(src1 = img_beach, alpha = 0.5 , src2 = img_desert,beta = 0.5 , gamma =0) # %50 , %50 


#show images

plt.figure()
plt.imshow(img_beach)

plt.figure()
plt.imshow(img_desert)

plt.figure()
plt.imshow(img_blended)
```

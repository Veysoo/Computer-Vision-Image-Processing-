MORPHOLOGICAL OPERATIONS

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

img = cv2.imread("image2.png",0)

plt.figure()
plt.axis("off")
plt.imshow(img,cmap="gray")
```

#### EROZYON

```python
kernel = np.ones((5,5),dtype= np.uint8)
result = cv2.erode(img,kernel,iterations=1)

plt.figure()
plt.axis("off")
plt.imshow(result,cmap="gray")
```

#### GENİŞLEME DILATION

```python
result2 = cv2.dilate(img,kernel,iterations=1)

plt.figure()
plt.axis("off")
plt.imshow(result2,cmap="gray")
```

####  AÇILMA (GÜRÜLTÜ AZALTMAK İÇİN)

```python
# gürültü ekliyoruz

whiteNoise = np.random.randint(0,2,size = img.shape[:2])
whiteNoise = whiteNoise*255


plt.figure()
plt.axis("off")
plt.imshow(whiteNoise,cmap="gray")

noisy_img = whiteNoise + img

plt.figure()
plt.axis("off")
plt.imshow(noisy_img,cmap="gray")
```

```python
# açılma yöntemi  (ilk erozyan  ikinci genişletme yapıyor )

opening = cv2.morphologyEx(noisy_img.astype(np.float32),cv2.MORPH_OPEN,kernel)

plt.figure()
plt.axis("off")
plt.imshow(opening,cmap="gray")
```

#### KAPATMA 

```python
# Black noise

blackNoise = np.random.randint(0,2,size = img.shape[:2])
blackNoise = blackNoise*(-255)

plt.figure()
plt.axis("off")
plt.imshow(blackNoise,cmap="gray")

noisy_img = blackNoise + img
noisy_img[noisy_img<= -245] = 0

plt.figure()
plt.axis("off")
plt.imshow(noisy_img,cmap="gray")
```

```python
# kapama yöntemi (ilk genişletme yapılıyor daha sonra erozyon yapılıyor)

close = cv2.morphologyEx(noisy_img.astype(np.float32),cv2.MORPH_CLOSE,kernel)

plt.figure()
plt.axis("off")
plt.imshow(close,cmap="gray")
```

####  GRADIENT 

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)

plt.figure()
plt.axis("off")
plt.imshow(gradient,cmap="gray")
```
